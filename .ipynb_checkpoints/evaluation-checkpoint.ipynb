# Friedman test

```python
from scipy.stats import friedmanchisquare

data1 = df_lv.groupby('station_id')['mae'].mean()
data2 = df_ha.groupby('station_id')['mae'].mean()
data3 = df_sarima.groupby('station_id')['mae'].mean()
data4= df_var_n5.groupby('station_id')['mae'].mean()
data5= df_var_d300.groupby('station_id')['mae'].mean()
data6= df_prophet.groupby('station_id')['mae'].mean()
# compare samples
stat, p = friedmanchisquare(data1,data2,data3,data4,data5,data6)
print('Statistics=%.3f, p=%.5f' % (stat, p))
# interpret
alpha = 0.05

if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

# Wilcoxon test

```python
def wilcoxon_test(m1,m2,error):
    model_1= m1.groupby('station_id')[error].mean()
    model_2= m2.groupby('station_id')[error].mean()
    merged_model=pd.merge(model_1,model_2,left_on='station_id',right_on='station_id')
    stat, p = wilcoxon(merged_model[error+'_x'], merged_model[error+'_y'], alternative='less')
    return p
#     print('Statistics=%.3f, p=%.20f' % (stat, p))
#     # interpret
#     bonferroni =8
#     alpha = 0.05
#     if p > (alpha/bonferroni):
#         print('Median of different is positive (fail to reject H0)')
#     else:
#         print('Median of different is negetive (reject H0)')
```

```python
print('P-value == 0.00625 and anything lower than that is significant')
print('           LV               HA')
print('SARIMA     %.3f            %.3f      ' % (wilcoxon_test(df_sarima,df_lv,'rmse'),wilcoxon_test(df_sarima,df_ha,'rmse')))
print('VAR_n=5    %.3f            %.3f      ' % (wilcoxon_test(df_var_n5,df_lv,'rmse'),wilcoxon_test(df_var_n5,df_ha,'rmse')))
print('VAR_d=300  %.3f            %.3f      ' % (wilcoxon_test(df_var_d300,df_lv,'rmse'),wilcoxon_test(df_var_d300,df_ha,'rmse')))
print('Prophet    %.3f            %.3f      ' % (wilcoxon_test(df_prophet,df_lv,'rmse'),wilcoxon_test(df_prophet,df_ha,'rmse')))
```

# Winner model

```python
from scipy.stats import friedmanchisquare


data3 = df_sarima.groupby('station_id')['mae'].mean()
data4= df_var_n5.groupby('station_id')['mae'].mean()
data5= df_var_d300.groupby('station_id')['mae'].mean()
# compare samples
stat, p = friedmanchisquare(data3,data4,data5)
print('Statistics=%.3f, p=%.5f' % (stat, p))
# interpret
alpha = 0.05

if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

```python
print('P-value == 0.016 and anything lower than that is significant')
print('          SARIMA         VAR_n=5')
print('SARIMA     -              -      ' )
print('VAR_n=5    %.3f           -      ' % (wilcoxon_test(df_var_n5,df_sarima,'rmse')))
print('VAR_d=300  %.3f           %.3f      ' % (wilcoxon_test(df_var_d300,df_sarima,'rmse'),wilcoxon_test(df_var_d300,df_var_n5,'rmse')))
```
